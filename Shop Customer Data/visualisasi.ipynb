2. Bar Chart

Bar chart dapat digunakan untuk membandingkan jumlah atau frekuensi data pada kategori yang berbeda. Dalam kasus ini, misalnya kita ingin melihat distribusi jenis kelamin dan profesi pada data.

```
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribusi Gender')
plt.show()
```

```
profession_counts = df['Profession'].value_counts()
plt.bar(profession_counts.index, profession_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Profession')
plt.ylabel('Count')
plt.title('Distribusi Profesi')
plt.show()
```

3.Box Plot

Box plot dapat digunakan untuk melihat distribusi data pada suatu variabel. Dalam kasus ini, misalnya kita ingin melihat distribusi umur pada data.

```
plt.boxplot(df['Age'])
plt.ylabel('Umur')
plt.title('Distribusi Umur')
plt.show()
```

4.Heatmap

Heatmap dapat digunakan untuk melihat hubungan antara dua atau lebih variabel pada data. Dalam kasus ini, misalnya kita ingin melihat hubungan antara semua variabel pada data.

```
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Variabel')
plt.show()
```
